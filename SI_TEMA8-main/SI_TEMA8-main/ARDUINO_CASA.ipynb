```python
import ipywidgets as widgets
from IPython.display import display
from datetime import datetime

# Função para registrar o evento junto com a data e hora e exibir o log em tempo real
def registrar_evento(evento):
    data_hora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_registros.append(f"{data_hora}: {evento}")
    label_status.value = f"{evento} (registrado em {data_hora})"
    atualizar_arquivo_log()

# Função para atualizar o arquivo de log
def atualizar_arquivo_log():
    with open("log.txt", "w") as arquivo:
        for registro in log_registros:
            arquivo.write(registro + "\n")

# Funções de controle dos dispositivos
def ligar_lampada(b):
    registrar_evento("Lâmpada ligada")

def desligar_lampada(b):
    registrar_evento("Lâmpada desligada")

def destrancar_porta(b):
    registrar_evento("Porta destrancada")

def trancar_porta(b):
    registrar_evento("Porta trancada")

def aumentar_temperatura(b):
    registrar_evento("Temperatura aumentada")

def diminuir_temperatura(b):
    registrar_evento("Temperatura diminuída")

# Criação dos botões para controlar os dispositivos
btn_lampada_on = widgets.Button(description="Ligar Lâmpada")
btn_lampada_off = widgets.Button(description="Desligar Lâmpada")
btn_porta_destrancar = widgets.Button(description="Destrancar Porta")
btn_porta_trancar = widgets.Button(description="Trancar Porta")
btn_aumentar_temperatura = widgets.Button(description="Aumentar Temperatura")
btn_diminuir_temperatura = widgets.Button(description="Diminuir Temperatura")

# Associação das funções aos eventos dos botões
btn_lampada_on.on_click(ligar_lampada)
btn_lampada_off.on_click(desligar_lampada)
btn_porta_destrancar.on_click(destrancar_porta)
btn_porta_trancar.on_click(trancar_porta)
btn_aumentar_temperatura.on_click(aumentar_temperatura)
btn_diminuir_temperatura.on_click(diminuir_temperatura)

# Criação de um rótulo para exibir o status dos dispositivos
label_status = widgets.Label(value="Status dos Dispositivos")

# Exibição dos widgets na interface
display(widgets.VBox([widgets.HBox([btn_lampada_on, btn_lampada_off]),
                      widgets.HBox([btn_porta_destrancar, btn_porta_trancar]),
                      widgets.HBox([btn_aumentar_temperatura, btn_diminuir_temperatura]),
                      label_status]))

# Inicializar o vetor de log
log_registros = []

# Carregar registros existentes do arquivo de log, se houver
try:
    with open("log.txt", "r") as arquivo:
        log_registros = arquivo.readlines()
except FileNotFoundError:
    pass
```
